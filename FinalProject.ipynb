```python
#transpose rows and columns
data_transp=data.set_index("Page").transpose().reset_index().rename(columns={"index":"Date"})
```

```python
#transpose rows and columns
data_transp.head()
```

```python
#reshape from wide to long format
#https://towardsdatascience.com/wide-to-long-data-how-and-when-to-use-pandas-melt-stack-and-wide-to-long-7c1e0f462a98
data_transp_long=pd.melt(data_transp,id_vars=["Date"],var_name="Web",value_name="Traffic")
```

```python
data_transp_long.head()
```

```python
#define a function to extract language from page name
import re #https://www.w3schools.com/python/python_regex.asp
def extract_lang(page):
    res= re.search("[a-z][a-z].wikipedia.org", page)
    if res:
        print(res)
        return res[0][0:2]
    return("na")

data_transp_long["language"]= data_transp_long.Web.map(extract_lang)
#the wikimedia pages fail the 
```

```python
print(data_transp_long["language"].unique())
```

wikimedia pages fail the regex search hence the na. Therefore, there are 7 languages plus the media language

```python
#A similar function to extract mode of access
def extract_mode(page): #https://www.w3schools.com/python/python_regex.asp
    res= re.search(".org_.+_", page)
    #print (res)
    if res:
        return res[0][5:]
    return("na")

data_transp_long["access_mode"]= data_transp_long.Web.map(extract_mode)
```

```python
#check for missing values
data_transp_long.isna().sum()
```

```python
#Remove rows and columns with null values
Final_data=data_transp_long.dropna()
```
